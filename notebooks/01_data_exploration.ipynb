```python
import pandas as pd
train_path = "train_FD001.txt"
test_path  = "test_FD001.txt"
rul_path   = "RUL_FD001.txt"

train_df = pd.read_csv(train_path, sep=" ", header=None)
test_df  = pd.read_csv(test_path, sep=" ", header=None)
rul_df   = pd.read_csv(rul_path, sep=" ", header=None)

train_df = train_df.iloc[:, :-2]
test_df  = test_df.iloc[:, :-2]

train_df.head()
```

```python
columns = [
    "unit", "time",
    "op_setting_1", "op_setting_2", "op_setting_3"
] + [f"sensor_{i}" for i in range(1,22)]

train_df.columns = columns
test_df.columns  = columns

train_df.head()
```

```python
columns = [
    "unit", "time",
    "op_setting_1", "op_setting_2", "op_setting_3"
] + [f"sensor_{i}" for i in range(1,22)]

train_df.columns = columns
test_df.columns  = columns

train_df.head()
```

```python
train_df.isnull().sum().sum()
```

```python
max_cycles = train_df.groupby("unit")["time"].max()
max_cycles.head()
```

```python
train_df["RUL"] = train_df.apply(
    lambda row: max_cycles[row["unit"]] - row["time"],
    axis=1
)

train_df[["unit", "time", "RUL"]].head()
```

```python
train_df["failure_24h"] = (train_df["RUL"] <= 24).astype(int)

train_df[["RUL", "failure_24h"]].tail(30)
```

```python
train_df["failure_24h"].value_counts(normalize=True)
```

```python
train_df.to_csv("train_processed.csv", index=False)
```
